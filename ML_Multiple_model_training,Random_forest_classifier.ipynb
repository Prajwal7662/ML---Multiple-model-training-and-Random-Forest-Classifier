```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# predict what is the time? if any one visiting -- is it lunch or dinner? time is the target variable

df.time.unique()
```

```python
# encoding,missing value treatment,scaling>> automate

df.info()
```

```python
df.isnull().sum()
```

```python
# categorical to numerical

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['time'] = encoder.fit_transform(df['time'])
df
```

```python
df['time'].unique()
```

```python
x = df.drop('time',axis=1)
y = df['time']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
# handling the missing values
# data encoding
# feature scaling

from sklearn.impute import SimpleImputer #missing value
from sklearn.preprocessing import OneHotEncoder #encoding
from sklearn.preprocessing import StandardScaler #scaling

from sklearn.pipeline import Pipeline #sequence of data transformer
from sklearn.compose import ColumnTransformer # groups all the pipeline steps for each of the col
```

```python
df
```

```python
df.sample(1)
```

```python
cat_cols = ['sex','smoker','day']
num_cols = ['total_bill','tip','size']
```

```python
# feature engineering automation using pipeline and col transformer

num_pipeline = Pipeline( steps=[ ('imputation',SimpleImputer(strategy='median')),
       ('scaling',StandardScaler())])

cat_pipeline = Pipeline(steps=[ ('imputation',SimpleImputer(strategy='most_frequent')),
       ('onehot',OneHotEncoder())])
```

```python
preprocessor = ColumnTransformer([('num_pipeline',num_pipeline,num_cols),
                   ('cat_pipeline',cat_pipeline,cat_cols)])
preprocessor
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Building the multiple models

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models = {'support vector classifier':SVC(),
          'DT classifier':DecisionTreeClassifier(),
          'Log reg':LogisticRegression()}
```

```python
models
```

```python
for i in range(len(models)):
  print(list(models.values())[i])
```

```python
from sklearn.metrics import accuracy_score

def model_train_eval(x_train,y_train,x_test,y_test,models):
  evaluation = {}
  for i in range (len(models)):
    model = list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    model_score = accuracy = accuracy_score(y_test,y_pred)
    evaluation[list(models.keys())[i]] = model_score
  return evaluation
```

```python
model_train_eval(x_train,y_train,x_test,y_test,models)
```
